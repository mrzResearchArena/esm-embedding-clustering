```python
superfamilies = []
for header, _ in esm.data.read_fasta('astral-SCOPe-K192.fa'):
    superfamily = '.'.join(header.split()[1].split('.')[:-1])
    superfamilies.append(superfamily)
#end-for
```

```python
superfamilies = list(dict.fromkeys(superfamilies))
print(superfamilies)
len(superfamilies)
```

```python
numericalClusters = [i for i in range(0, len(superfamilies))]
```

```python
KeysNumericalValuesAlphabetical = dict(zip(numericalClusters, superfamilies))
```

```python
KeysAlphabeticalValuesNumerical = dict(zip(superfamilies, numericalClusters))
```

```python
# KeysNumericalValuesAlphabetical # {0: 'a.1.1', 1: 'a.2.11', 2: 'a.3.1',...}
```

```python
# KeysAlphabeticalValuesNumerical # {'a.1.1': 0, 'a.2.11': 1, 'a.3.1': 2,...}
```

```python
def numericalToAlphabetical(Cs):
    clusters = []
    for c in Cs:
        clusters.append(KeysNumericalValuesAlphabetical[c])
    #end-for
    return clusters
#end-def

# numericalToAlphabetical(v)
```

```python
v = np.load('PyTorchKMeans+ESM1b.npy')
D['KMeans+ESM1b'] = numericalToAlphabetical(v)
D
```

```python
v = np.load('HC+ESM1b.npy')
D['hierarchical+ESM1b'] = numericalToAlphabetical(v)
D
```

```python
v = np.load('HC+ESM1v.npy')
D['hierarchical+ESM1v'] = numericalToAlphabetical(v)
D
```
