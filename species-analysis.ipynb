```python
import esm

import pandas as pd
import numpy as np

import re # regular expression (It helps to find the pattern from sequence header.) 
```

```python
dataset = '/home/rafsanjani/backup-research/esm/astral-scopedom-seqres-sel-gs-bib-95-2.08.fa'
```

```python
C=1

for header, sequence in esm.data.read_fasta(dataset):
    # print(header)
    
    v = re.findall(pattern='\{.*}', string=header)[0]
    # print(v)
    # v1 = '{Bacillus subtilis [TaxId: 1423]}'
    # filename = 'Example_file_[extra_descriptor].txt'
    
    p = re.compile(r'\[.*\]')
    
    v = re.sub(p, '', v)
    v = v[1:-1].strip()
    
    print(f'{C}: {v}')
    
    # break
    # if C==100: break
    C+=1
    # break
```

```python
C=1

organism = 0
for header, sequence in esm.data.read_fasta(dataset):
    # print(header)
    
    v1 = re.findall(pattern='\{.*}', string=header)[0]
    # print(v1)
    
    # v1 = '{Bacillus subtilis [TaxId: 1423]}'
    # filename = "Example_file_[extra_descriptor].ext"
    p = re.compile(r'\[.*\]')
    v2 = re.sub(p, '', v1)
    v2 = v2[1:-1].strip()
    
    v3 = v2.split()[0]
    
    if v3 == 'Human':
        organism += 1
    
    # print(f'{C}: {v2.split()[0]}')
    
    # break
    if C==1000: break
    C+=1
    # break
print(organism)
```
